```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import pandas as pd

csv_15 = pd.read_csv('./players_15.csv')

#input_columns = ['overall', 'age', 'height_cm', 'weight_kg', 'international_reputation', 'weak_foot', 'skill_moves']
input_columns = ['overall']

x_train = csv_15[input_columns].to_numpy()
y_train = csv_15['potential'].to_numpy()

x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
```

```python
dct_regr_5 = DecisionTreeRegressor(max_depth=5)
dct_regr_5.fit(x_train, y_train)
pred_5 = dct_regr_5.predict(x_train)

dct_regr_10 = DecisionTreeRegressor(max_depth=10)
dct_regr_10.fit(x_train, y_train)
pred_10 = dct_regr_10.predict(x_train)
```

```python
plt.figure()
plt.scatter(x_train, y_train, s=20, edgecolor="black", c="green", label="data")
plt.plot(x_train, pred_5, color="red", label="max_depth=5", linewidth=2)
plt.plot(x_train, pred_10, color="blue", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
